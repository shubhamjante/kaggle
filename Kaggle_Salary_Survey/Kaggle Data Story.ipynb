# Let's see who make more money, men or women?

```python
gender_palette = ['pink','#95d0fc']
sns.set_palette(gender_palette)
data_EDA['Q9'] = data_EDA['Q9'].apply(int)
px2 = data_EDA[data_EDA['Q1'].isin(['Male', 'Female'])].groupby(['Q1'])['Q9'].mean().sort_values().plot(kind='bar',
                                                                                                       figsize=(16,7))
px2.set(xlabel='Gender', ylabel='Average Yearly Compensation (USD)')
```

# How does your current location affect your pay?

```python
# Top 20 Respondents Location/Country
sns.set_palette('husl')
px3 = data_EDA[data_EDA['Q3'] != 'Other']['Q3'].value_counts().nlargest(20).plot(kind='bar', figsize=(16,7))
px3.set_xticklabels(px3.get_xticklabels(), rotation=90)
```

```python
# Top 10 countries with highest income
sns.set_palette(sns.light_palette("green",reverse=True, n_colors=10))
px4 = data_EDA.groupby(['Q3'])['Q9'].mean().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(16,7))
px4.set(xlabel='Country', ylabel='Average Yearly Compensation (USD)')
```

```python
# Education Level Distribution
sns.set_palette('husl')
px5 = data_EDA['Q4'].value_counts(normalize=True).plot(kind='pie', autopct='%.1f%%', figsize=(16,7))
px5.axis('equal')
px5.get_yaxis().set_visible(False)
```
